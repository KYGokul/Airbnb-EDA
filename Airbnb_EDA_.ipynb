#### Chart - 4 - Highest and Lowest AirBnb prices by location

````python
# Chart - 4 visualization code
# Getting the highest prices by the location
highest_prices = new_df.groupby('neighbourhood_group',as_index=False)['price'].max().sort_values(['price'],ascending = False).rename(columns = {'price':'Highest price','neighbourhood_group':'Location'})

# Getting the lowest prices by the location
lowest_prices = new_df.groupby('neighbourhood_group',as_index=False)['price'].min().sort_values(['price'],ascending = True).rename(columns = {'price':'Lowest price','neighbourhood_group':'Location'})
````

````python
# Merging the high prices and low prices by their location
lowest_highest_prices = pd.merge(highest_prices, lowest_prices, on = 'Location')
lowest_highest_prices
````

````python
# Set up the figure and axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot the highest prices
ax1.bar(lowest_highest_prices['Location'], lowest_highest_prices['Highest price'], color='b', alpha=0.7)
ax1.set_title('Highest Airbnb Prices by Location')
ax1.set_xlabel('Location')
ax1.set_ylabel('Price')
ax1.set_ylim(0, 11000)  # Set y-axis limit for better readability

# Add bar values above the bars for highest prices
for i, value in enumerate(lowest_highest_prices['Highest price']):
    ax1.text(i, value + 200, str(value), ha='center', va='bottom')

# Plot the lowest prices
ax2.bar(lowest_highest_prices['Location'], lowest_highest_prices['Lowest price'], color='r', alpha=0.7)
ax2.set_title('Lowest Airbnb Prices by Location')
ax2.set_xlabel('Location')
ax2.set_ylabel('Price')
ax2.set_ylim(0, 20)  # Set y-axis limit for better readability

# Add bar values above the bars for lowest prices
for i, value in enumerate(lowest_highest_prices['Lowest price']):
    ax2.text(i, value + 1, str(value), ha='center', va='bottom')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

Bar chart is ideal for straighforward comparision of the highest prices and lowest prices between different neighbourhood.

##### 2. What is/are the insight(s) found from the chart?

* Brooklyn, Queens, Manhattan locations have the highest price listings of $1000.

* Bronx and Staten Island is highest price listings are less than $1000.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

In Brooklyn, Queens and Manhattan the lowest price is $10 where as Staten Island is 13 and Bronx has 20 dollars.

#### Chart - 10 - Top 10 Hosts by Number of Listings

````python
# Chart - 11 visualization code
# Find the host with the maximum count of listings
host_counts = new_df['host_name'].value_counts()
host_counts_df = pd.DataFrame({'host_name': host_counts.index, 'listing_count': host_counts.values})
sorted_host_counts_df = host_counts_df.sort_values(by='listing_count', ascending=False)
sorted_host_counts_df.head(5)
````

#### Chart - 16 - Average prices of different room types

````python
# Group by 'room_type' and calculate the mean price for each room type
room_type_prices = new_df.groupby('room_type')['price'].mean().reset_index()

# Create a colormap
cmap = plt.get_cmap('tab10')

# Create an array of colors from the colormap
colors = cmap(np.linspace(0, 1, len(room_type_prices)))

# Create a bar chart with custom colors
plt.figure(figsize=(10, 6))
bars = plt.bar(room_type_prices['room_type'], room_type_prices['price'], color=colors)

plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.title('Average Price Across Room Types')

# Adding annotations
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
````
